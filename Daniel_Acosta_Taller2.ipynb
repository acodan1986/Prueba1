Para realizar esta visualización se tomo un archivo de tipo .edges que datos tomados de la corteza viusual en ratones, el lugar desde donde pueden descargarse estas bases es el siguiente: http://networkrepository.com/bn-mouse-visual-cortex-2.php.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
from pyvis.network import Network
import matplotlib.pyplot as plt
import csv
```

```python
cortex = nx.read_edgelist('bn-mouse_visual-cortex_2.edges',create_using=nx.Graph(),nodetype=int)


print(nx.info(cortex))
print(nx.is_directed(cortex))
```

De la información descriptiva se puede resaltar que este es un grafo no dirigido, y que en promedio cada nodo tiene 2.21 enlaces.

```python

print("Número de nodos:", cortex.number_of_nodes())
print("Número de enlaces:", cortex.number_of_edges())
```

```python
list(cortex.nodes)
```

```python
list(cortex.edges)
```

```python
plt.figure(figsize=(50,20))
plt.subplot(121)
nx.draw(cortex, with_labels=True)
```

### ABSTRACCIÓN

####  QUE?

* Se trata de una grafo en forma de red, el cual pretende mostrar las conexiones existentes en diferentes partes de la corteza visual en ratones. En este grafico se muestra solamente el vinculo no dirigido ni cuantificado de la relación existente entre las diversas zonas de la corteza visual. Las zonas (nodos) si presentan una codificación numerica que no es cuantificable. El tipo de data set es una lista de enlaces.

#### PORQUE ?

* La vizualización tiene por objetivo Presentar, ademas de poder realizar un Locate de los nodos en que estemos interesados. Entre los objetivos se tiene organizar mediante una topologia las relaciones existentes entre los nodos.

#### COMO?

* El verbo clave en esta visualización es el ordenar de manera espacial los nodos y aristas para identificar estructuras de organización y agrupamiento entre las diversas zonas de la corteza visual.

### MARCAS Y CANALES

* Se utilizan marcas de puntos para los nodos y marcas de lineas para los enlaces.
* La posición espacial juega un papel importante aqui para poder organizar mejor los nodos y permitir distinguir relaciones entre ellos.
* El canal del color se usa para resaltar los nodos.

### CUMPLIMIENTO DE REGLAS

* El uso del 2D esta muy bien justificado ya que de haberse presentado en 1D (Lista de enlaces) no seria posible realizar insights facilmente.
* Primero la función luego la forma: Esta visualización cumple con esta regla ya que presenta la información de manera sencilla.

### MEJORAS POSIBLES

* En general esta vizualización me parece buena, sin embargo buscaria mejorar la organizacion espacial de los nodos para evitar problemas de oclusión, asi mismo podria utilizar lineas curvas para las aristas para buscar que estas no pasen sobre nodos con los que no hagan conexión.

* Asi mismo si el dataset contemplara una variable cualitativa que midiera "la fuerza" de la conexión usaria un canal de color y luminancia para resaltar esas relaciones.
